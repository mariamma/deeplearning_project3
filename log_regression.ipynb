```python
def findTfIdfTrainSet(df_rowcount, row_num, X_tfidf, feature_names):
    sent1_index = 0 + row_num
    sent2_index = df_rowcount + row_num
    tfidfdict = {}
    
    doc = sent1_index
    feature_index = X_tfidf[doc,:].nonzero()[1]
    tfidf_scores = zip(feature_index, [X_tfidf[doc, x] for x in feature_index])
    for w, s in [(feature_names[i], s) for (i, s) in tfidf_scores]:
        tfidfdict[w]=s
        
    doc = sent2_index
    feature_index = X_tfidf[doc,:].nonzero()[1]
    tfidf_scores = zip(feature_index, [X_tfidf[doc, x] for x in feature_index])
    for w, s in [(feature_names[i], s) for (i, s) in tfidf_scores]:
        tfidfdict[w]=s    
    return tfidfdict   
```

```python
def findSentenceCorrelation(df_rowcount, row, model_w2v2, X_tfidf, feature_names,word_vectors):
    #print(row.name)
    tfidf_dict = findTfIdfTrainSet(df_rowcount, row.name,X_tfidf, feature_names)
    #print(tfidf_dict)
    sent1 = tokenize_sentence(row['sentence1'])
    sent2 = tokenize_sentence(row['sentence2'])
    #print(sent1, sent2)
    sim = 0
    for word1 in sent1:
        for word2 in sent2:
            if word1 in word_vectors.vocab:
                if word2 in word_vectors.vocab:
                    #print("Row: ", row.name, " word1: ", word1, " word2: ", word2, " sim_score: ", model_w2v2.similarity(word1, word2), " tfidf : ",  tfidf_dict[word1], " | " ,tfidf_dict[word2])
                    sim = sim + model_w2v2.similarity(word1, word2)*tfidf_dict[word1]*tfidf_dict[word2]
    return sim                        
                    
```

```python
def findTrainWeightedVecRepresentation(train_rowcount,row, model_w2v2, X_tfidf, word_vectors, feature_names):
    tfidf_dict = findTfIdfTrainSet(train_rowcount,row.name,X_tfidf, feature_names)
    sent1 = tokenize_sentence(row['sentence1'])
    sent2 = tokenize_sentence(row['sentence2'])
    sim = 0                    
    bucket1 = [0] * 300                              
    bucket2 = [0] * 300                                                            
    for word1 in sent1:
        vec1 = [0] * 300                                                      
        if word1 in word_vectors.vocab:
            vec1 =  model_w2v2[word1]
            for j in range(300):
                bucket1[j] = bucket1[j] + tfidf_dict[word1] * vec1[j]
    for word1 in sent2:
        vec1 = [0] * 300                                                      
        if word1 in word_vectors.vocab:
            vec1 =  model_w2v2[word1]  
            for j in range(300):
                bucket2[j] = bucket2[j] + tfidf_dict[word1] * vec1[j]                         
    return bucket1, bucket2                        
    
```

```python
word_vectors = model_w2v2.wv
feature_names = tfidf.get_feature_names()
train_rowcount = traindf.shape[0]
for i, row in traindf.iterrows():
    bucket1, bucket2 = findTrainWeightedVecRepresentation(train_rowcount,row, model_w2v2, X_tfidf, word_vectors, feature_names)
    for j in range(0,300,1):
        traindf.at[i,'{}_vec1'.format(j)] = bucket1[j]
        traindf.at[i,'{}_vec2'.format(j)] = bucket2[j]                                             
```

```python
def encodeY(val):
    if(val.strip() == "entailment"):
        return 0
    elif (val.strip() == "neutral"):
        return 1
    elif (val.strip() == "contradiction"):
        return 2
    else:
        return 3
```
